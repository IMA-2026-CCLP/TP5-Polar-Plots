```python
import numpy as np
import plotly.graph_objects as go
```

```python
# 1. Definir la resolución de la malla esférica
theta = np.linspace(0, np.pi, 100)      # Ángulo polar (de 0 a 180 grados, eje Z)
phi = np.linspace(0, 2 * np.pi, 100)    # Ángulo azimutal (de 0 a 360 grados, plano XY)
theta, phi = np.meshgrid(theta, phi)
```

```python
R = np.abs(np.cos(theta)**3)  # SPL iría acá

# 3. Convertir de coordenadas esféricas a cartesianas (X, Y, Z) para graficar
X = R * np.sin(theta) * np.cos(phi)
Y = R * np.sin(theta) * np.sin(phi)
Z = R * np.cos(theta)
```

```python
# 4. Crear la superficie 3D (Balloon Plot)
fig = go.Figure(data=[go.Surface(
    x=X, y=Y, z=Z,
    surfacecolor=R,        # El color representará la magnitud de la radiación (R)
    colorscale='Jet',      # Escala de colores típica para directividad
    showscale=True,
    colorbar=dict(title="Magnitud")
)])

# 5. Configurar el diseño para que el espacio 3D sea simétrico
fig.update_layout(
    title='Directividad de la Fuente (Balloon Plot)',
    scene=dict(
        xaxis=dict(title='Eje X', range=[-1, 1]),
        yaxis=dict(title='Eje Y', range=[-1, 1]),
        zaxis=dict(title='Eje Z', range=[-1, 1]),
        aspectratio=dict(x=1, y=1, z=1) # Fundamental para que el globo no se deforme
    ),
    margin=dict(l=0, r=0, b=0, t=50) # Ajustar márgenes
)

# Mostrar el gráfico interactivo (Se abre en tu navegador o entorno Jupyter)
fig.show()
```
